```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
suppose you have to represent two curves in one plot. Inorder to differentiate them, we can use plt.legends() to describe
the said curves and tell us which color represents what curve in simple terms.
"""

x = [1,2,3,4]
y1 = [1,2,3,4]
y2 = [4,5,6,7]

plt.plot(x,y1)
plt.plot(x,y2)

"""
here the first list indicates the labels assigned to each curve. so orange curve is y1 and blue is y2.
fontsize changes the size of the legend while if shadow = bool, it adds shadow effect to the legend. 
By default shadow is None.

"""

plt.legend(['y1','y2'],loc="lower right",fontsize=15,shadow=bool)
```
